# Cálculo índice de salud de los gases


La detección de ciertos gases generados en un transformador lleno de aceite en servicio es,
con frecuencia, la primera indicación disponible de un mal funcionamiento que, si no se corrige,
puede conducir a la falla. Arcos, descargas parciales, chispas de baja energía, sobrecarga severa,
falla del motor de la bomba y sobrecalentamiento en el sistema de aislamiento son algunos de los
posibles mecanismos. Estas condiciones que ocurren individualmente, o como varios eventos
simultáneos, pueden resultar en la descomposición de los materiales aislantes y la formación de
varios gases combustibles y no combustibles.

# Implementación de función

```python
def gases(H2,CH4,C2H6,C2H4,C2H2,CO,CO2):
  #Para el gas H2
  if H2 <= 100:
    valorg_1 = 1 * 2
  elif 100 < H2 <= 200 :
    valorg_1 = 2 * 2
  elif 200 < H2 <= 300 :
    valorg_1 = 3 * 2
  elif 300 < H2 <= 500 :
    valorg_1 = 4 * 2
  elif 500 < H2 <= 700 :
    valorg_1 = 5 * 2
  else :
    valorg_1 = 6 * 2 
  # Para el gas CH4
  if CH4 <= 75:
    valorg_2 = 1 * 3
  elif 75 < CH4 <= 125 :
    valorg_2 = 2 * 3
  elif 125 < CH4 <= 200 :
    valorg_2 = 3 * 3
  elif 200 < CH4 <= 400 :
    valorg_2 = 4 * 3
  elif 400 < CH4 <= 600 :
    valorg_2 = 5 * 3
  else :
    valorg_2 = 6 * 3 

  # Para el gas C2H6
  if C2H6 <= 65:
    valorg_3 = 1 * 3
  elif 65 < C2H6 <= 80 :
    valorg_3 = 2 * 3
  elif 80 < C2H6 <= 100 :
    valorg_3 = 3 * 3
  elif 100 < C2H6 <= 120 :
    valorg_3 = 4 * 3
  elif 120 < C2H6 <= 150 :
    valorg_3 = 5 * 3
  else :
    valorg_3 = 6 * 3 

  # Para el gas C2H4

  if C2H4 <= 50:
    valorg_4 = 1 * 3
  elif 50 < C2H4 <= 80 :
    valorg_4 = 2 * 3
  elif 80 < C2H4 <= 100 :
    valorg_4 = 3 * 3
  elif 100 < C2H4 <= 150 :
    valorg_4 = 4 * 3
  elif 150 < C2H4 <= 200 :
    valorg_4 = 5 * 3
  else :
    valorg_4 = 6 * 3 
  
  # Para el gas C2H2

  if C2H2 <= 3:
    valorg_5 = 1 * 5
  elif 3 < C2H2 <= 7 :
    valorg_5 = 2 * 5
  elif 7 < C2H2 <= 35 :
    valorg_5 = 3 * 5
  elif 35 < C2H2 <= 50 :
    valorg_5 = 4 * 5
  elif 50 < C2H2 <= 80 :
    valorg_5 = 5 * 5
  else :
    valorg_5 = 6 * 5

  # Para el gas CO

  if CO <= 350:
    valorg_6 = 1 * 1
  elif 350 < CO <= 700 :
    valorg_6 = 2 * 1
  elif 700 < CO <= 900 :
    valorg_6 = 3 * 1
  elif 900 < CO <= 1100 :
    valorg_6 = 4 * 1
  elif 1100 < CO <= 1400 :
    valorg_6 = 5 * 1
  else :
    valorg_6 = 6 * 1 

  # Para el gas CO2

  if CO2 <= 2500:
    valorg_7 = 1 * 1
  elif 2500 < CO2 <= 3000 :
    valorg_7 = 2 * 1
  elif 3000 < CO2 <= 4000 :
    valorg_7 = 3 * 1
  elif 4000 < CO2 <= 5000 :
    valorg_7 = 4 * 1
  elif 5000 < CO2 <= 7000 :
    valorg_7 = 5 * 1
  else :
    valorg_7 = 6 * 1 
  # Se calcula el índice de gases disueltos en el aceite.   
  DGAF = (valorg_1 + valorg_2 + valorg_3 + valorg_4 + valorg_5 + valorg_6 + valorg_7) / (18)
  #(valorg_1,valorg_2,valorg_3,valorg_4,valorg_5,valorg_6,valorg_7)
  HIDGAF = 100 - (((DGAF - 1) * 100) / 5)

  # Se halla la condición del índice de salud referente a los gases

  if DGAF < 1.2 :
    aviso = "El transformador se encuentra en buenas condiciones"
  elif 1.2 <= DGAF < 1.5 :
    aviso = "El transformador se encuentra en condiciones aceptables"
  elif 1.5 <= DGAF < 2 :
    aviso = "Precaución"
  elif 2 <= DGAF < 3 :
    aviso = "El transformador se encuentra en condiciones pobres "
  else:
    aviso = "El transformador se encuentra en condiciones muy pobres"
  return (f"El índice de salud de gases disueltos del transformador es: {DGAF}. {aviso}. Índice de salud (HI) del transformador referente a los gases: {HIDGAF}%")
```

Se utiliza la función de gases para calcular el índice de salud. A continuación se deben ingresar los siguientes parámetros de gases separados por una coma en el siguiente orden.

Orden de gases: H2,CH4,C2H6,C2H4,C2H2,CO,CO2



```python
gases(1800,1000,35,200,1,350,10000)
```
